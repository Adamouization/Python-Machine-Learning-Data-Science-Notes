# Line Charts

```python
df = pd.read_csv("data/coronavirus-morocco/corona_morocco.csv").dropna()
df['Date'] = pd.to_datetime(df["Date"], utc=True, infer_datetime_format=True)
df = df.set_index("Date")
df.tail(5)
```

Use `sns.lineplot`. Can plot individual columns of the data and pass a label to tell the difference:

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Set title
plt.title("Number of confirmed COVID-19 cases in Morocco")

sns.lineplot(data=df["Confirmed"], label="Confirmed")
sns.lineplot(data=df["Deaths"], label="Deaths")
sns.lineplot(data=df["Recovered"], label="Recovered")

# Add label for horizontal axis
plt.xlabel("Dates")
```

# Bar Charts and Heatmaps

```python
df = pd.read_csv("data/us-weather-history/KHOU.csv")
df = df.set_index("date")

# Select first 30 dates (there are actually 365 rows, won't fit in the bar chart's x axis)
df = df.iloc[:30, :]
df.head(3)
```

## Bar chart

Use `sns.barplot`:
* `x` is the category to separate (in this case the days that are in the index).
* `y` is the actual data that will determine the height of each bar.

```python
plt.figure(figsize=(19, 5))
sns.barplot(x=df.index, y=df["actual_mean_temp"])

# Rotate x-axis labels and move them to the right
plt.xticks(rotation=45, ha='right')
```

## Heatmap

```python
df = pd.read_csv("data/us-weather-history/KHOU.csv")
df = df.set_index("date")

# Select first 30 dates (there are actually 365 rows, won't fit in the bar chart's x axis)
df = df.iloc[:14, :].loc[:, ["actual_mean_temp", "actual_min_temp", "actual_max_temp", "average_min_temp", "average_max_temp"]]
df
```

Use `sns.heatmap`:
* `annot=True` to show values in each cell of the map.

```python
plt.figure(figsize=(12,7))
plt.title("Average and Actual mean/min/max temperatures in KHOU during the first 2 weeks of July 2014")
sns.heatmap(data=df, annot=True) 

# Add label for axes
plt.xlabel("Temperature")
plt.ylabel("Date")
```

# Scatter Plots

## Single-colour scatter plots + regression lines

```python
df = pd.read_csv("data/Medical Cost Personal Datasets/insurance.csv")
df = df.rename(columns={"bmi":"Body Mass Index"})
df.head()
```

Use `sns.scatterplot`:
* Specify the two variables in `x` and `y` (usually `y` is the value to predict)

--> useful for detecting linear correlations between two variables e.g. the plot below shows a positive correlation between the BMI and the higher insurance costs, which makes sense since high BMI is typically associated with higher risk of chronic disease.

```python
sns.scatterplot(x=df["Body Mass Index"], y=df["charges"])
```

Can add a regression line (line that best fits the data) to verify the pattern observed. Do this using `sns.regplot` instead:

```python
sns.regplot(x=df["Body Mass Index"], y=df["charges"])
```

## Coloured scatter plots

With one colour (above), can only show relationship between 2 variables. By adding more colours, can use scatter plots to display the relationships between 3 variables.

Example: To understand how smoking affects the relationship between BMI and insurance costs, color-code the points by 'smoker', and plot the other two columns ('BMI', 'charges') on the axes:
* Use `sns.scatterplot`
* Specify the variable to colour-code with `hue` (for now convert `.to_list()` due to bug in seaborn)

--> Useful to show that smokers pay much more with increasing BMI than non-smokers.

```python
sns.scatterplot(data=df, x="Body Mass Index", y="charges", hue=df["smoker"].to_list())
```

Plot 2 regression lines using `sns.lmplot`, showing that the slope is much steeper for smokers than non-smokers:
* specify dataset in `data`
* pass individual column names for `x`, `y` and `hue`

```python
sns.lmplot(data=df, x="Body Mass Index", y="charges", hue="smoker")
```

## Categorical scatter plots

Types of data:
* BMI is numerical data (can be plotted in a scatter plot to show relationship)
* smoker categorical data. 

To plot categorical data like smoker, can use a swarm plot with `sns.swarmplot`:

```python
sns.swarmplot(data=df, x="smoker", y="charges")
```

This shows that on average, non-smokers are charged less than smokers, and the customers who pay the most are smokers whereas the customers who pay the least are non-smokers.


# Distributions

```python
df = pd.read_csv("data/iris-species/Iris.csv")
df
```

## Histograms

Histogram to show the variation of petal length in iris flowers. 

Use `sns.distplot`:
* `a` is for the column to plot
* `kde`: to plot a gaussian kernel density estimate or not

```python
plt.figure(figsize=(8, 5))
sns.distplot(a=df["PetalLengthCm"], kde=False)

# Style
plt.title("Histogram of iris flowers petal length in cm")
plt.xlabel("Petal Length (cm)")
plt.ylabel("# Iris Flowers")
```

## Density Plots (KDE)

Kernel density estimate (KDE) plots are like smoothed histograms. 

Use `sns.kdeplot`:
* `data` for the column to plot
* `shaded` to colour the area under the curve

```python
sns.kdeplot(data=df["PetalLengthCm"], shade=True)
```

## 2D KDE Plots

Create a 2D KDE plot with `sns.jointplot`:
* `x` and `y` for the 2 distributions to plot
* `kind` for type of 2D distribution (can be `kde`, `scatter` or `reg`)

```python
sns.jointplot(x=df["PetalLengthCm"], y=df["SepalWidthCm"], kind="kde")
sns.jointplot(x=df["PetalLengthCm"], y=df["SepalWidthCm"], kind="scatter")
```

## Colour-coded distributions

Separate different types of iris flowers into unique DFs:

```python
print(f"Type of iris flowers: {df['Species'].unique()}")
df_setosa = df.loc[df["Species"] == "Iris-setosa"]
df_versicolor = df.loc[df["Species"] == "Iris-versicolor"]
df_virginica = df.loc[df["Species"] == "Iris-virginica"]
```

Create a different histogram for each iris flower species. 

Use `sns.distplot` for each species and apply a `label` to tell them apart:

```python
sns.distplot(a=df_setosa["PetalLengthCm"], label="setosa", kde=False)
sns.distplot(a=df_versicolor["PetalLengthCm"], label="versicolor", kde=False)
sns.distplot(a=df_virginica["PetalLengthCm"], label="virginica", kde=False)

# Style
plt.title("Histogram of petal lengths by species")
plt.xlabel("Petal Length (cm)")
plt.legend()  # Force legend to appear using specified labels
```

Can do the same for KDE plots:

```python
sns.kdeplot(data=df_setosa["PetalLengthCm"], label="setosa", shade=True)
sns.kdeplot(data=df_versicolor["PetalLengthCm"], label="versicolor", shade=True)
sns.kdeplot(data=df_virginica["PetalLengthCm"], label="virginica", shade=True)

# Style
plt.title("Distribution of petal lengths by species")
plt.xlabel("Petal Length (cm)")
plt.legend()  # Force legend to appear using specified labels
```

# Summary

![](https://i.imgur.com/2VmgDnF.png)

3 categories of charts seen:

* **Trends** - A trend is defined as a pattern of change.
    * `sns.lineplot` - Line charts are best to show trends over a period of time, and multiple lines can be used to show trends in more than one group.
    

* **Relationship** - There are many different chart types that you can use to understand relationships between variables in your data.
    * `sns.barplot` - Bar charts are useful for comparing quantities corresponding to different groups.
    * `sns.heatmap` - Heatmaps can be used to find color-coded patterns in tables of numbers.
    * `sns.scatterplot` - Scatter plots show the relationship between two continuous variables; if color-coded, we can also show the relationship with a third categorical variable.
    * `sns.regplot` - Including a regression line in the scatter plot makes it easier to see any linear relationship between two variables.
    * `sns.lmplot` - This command is useful for drawing multiple regression lines, if the scatter plot contains multiple, color-coded groups.
    * `sns.swarmplot` - Categorical scatter plots show the relationship between a continuous variable and a categorical variable.


* **Distribution** - We visualize distributions to show the possible values that we can expect to see in a variable, along with how likely they are.
    * `sns.distplot` - Histograms show the distribution of a single numerical variable.
    * `sns.kdeplot` - KDE plots (or 2D KDE plots) show an estimated, smooth distribution of a single numerical variable (or two numerical variables).
    * `sns.jointplot` - This command is useful for simultaneously displaying a 2D KDE plot with the corresponding KDE plots for each individual variable.
